```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv("C:\\Users\\Zachary_B\\Desktop\\data sci bootcamp\\Data Visualization\\data set\\bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9, 6))
plt.barh(y = df_used_cars["Brand"],
         width = df_used_cars["Cars Listings"],
        color = "rgbwymc")  
plt.show()
```

```python
plt.figure(figsize=(9, 6))
plt.barh(y = df_used_cars["Brand"],
       width = df_used_cars["Cars Listings"],
       color = "grymwck")
plt.xticks(fontsize = 13)  
plt.yticks(fontsize = 13)
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold")
plt.xlabel("Number of Listinga", fontsize = 13)
plt.show()
plt.savefig("Used_Cars_Horizontal_Bar.png")
```
