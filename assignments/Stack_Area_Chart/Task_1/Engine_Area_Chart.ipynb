```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_types = pd.read_csv("C:\\Users\\Zachary_B\\Desktop\\data sci bootcamp\\Data Visualization\\data set\\stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
colors = ["#262626", "#8B0A50", "#2F4F4F", "#8FBC8F"]
labels = ["Diesel", "Petrol", "Gas", "Other"]
sns.set_style("white")
plt.figure(figsize = (12, 6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              df_fuel_engine_types["Other"],
              colors = colors,
             edgecolor = 'none')
plt.xticks(df_fuel_engine_types["Year"], rotation = 45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number of Cars", fontsize = 13)
plt.title("Engine Fuel Types (1982 - 2016)", fontsize = 14, weight = "bold")
sns.despine()
plt.show()
```
